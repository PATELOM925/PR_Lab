Use the Pima Indians diabetes dataset which is often used for binary classification problem. 
Do the following using suitable python library using this dataset.

1. Since the attributes present in this dataset is of different scales, first rescale 
     the attributes of this dataset so that all the attributes have the same scale. Use the scale 0 to 1 for the feature information of this dataset. 
2. The next is to binarize the data. Transform the whole data set using a binary threshold. 
     All values above the threshold are marked 1 and all equal to or below are marked as 0. 
     Use the threshold value 0.0 for all the features of this dataset. Print the head of the transformed data.

```python
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df1 = df.drop(columns='Outcome')
df1
```

#### 1) Scale Features

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
df1_rescaled = mm.fit_transform(df1)
df2 = pd.DataFrame(df1_rescaled, columns=df1.columns)
df2
```

#### 2) Binarize data

```python
df2.describe()
```

```python
from sklearn.preprocessing import Binarizer

thresholds = {
    'Pregnancies': 0.0,
    'Glucose': 0.5,
    'BloodPressure': 0.56,
    'SkinThickness': 0.2,
    'Insulin': 0.09,
    'BMI': 0.47,
    'DiabetesPedigreeFunction': 0.3,
    'Age': 0.5
}

df3 = pd.DataFrame()
for column in df2.columns:
    threshold = thresholds[column]
    binarizer = Binarizer(threshold=threshold)
    binarized_column = binarizer.fit_transform(df2[[column]])
    df3[column] = binarized_column.flatten()

df3
```

#### 3 Standarize Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()

Standardize_data = ss.fit_transform(df3)
df4 = pd.DataFrame(Standardize_data, columns=df3.columns)
df4
```

# Q2

    Perform the following cleaning operations on the spam.csv dataset.
    1.Remove the punctuations (e.g., !, #, @ etc.). Make use of apply() 
    function or use the regular expression to do the same. 
    2.Remove the stopwords following the library available in nltk.


```python
import string
import nltk
from nltk.corpus import stopwords
```

```python
spam = pd.read_csv('spam.csv',encoding='ISO-8859-1')
spam
```

```python
spam.isna().sum()
```

```python
spam.describe()
```

```python
spam.drop(columns='v1')
```

#### 1) Remove the punctuations (e.g., !, #, @ etc.)

```python
def rem_punct(text):
    s = text
    return s.translate(str.maketrans('', '', string.punctuation))

spam['v2'] = spam['v2'].apply(rem_punct)
spam['v2']
```

#### 2) Remove the stopwords using nltk library

```python
def rem_sw(text):
    eng = stopwords.words('english')
    stopWords = set(eng)
    words = text.split()
    not_stop_words = [word for word in words if word.lower() not in stopWords]
    return ' '.join(not_stop_words)

spam['v2'] = spam['v2'].apply(rem_sw)
spam['v2']
```
