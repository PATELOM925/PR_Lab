```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Q1

```python
df0 = load_iris()
```

```python
df0
```

```python
df = pd.DataFrame(data=df0.data, columns=df0.feature_names)
df['target'] = df0.target
df
```

### a)

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
data_ss = ss.fit_transform(df0.data)

# Convert to DataFrame for better visibility
df1 = pd.DataFrame(data_ss, columns=df0.feature_names)
df1.head()
```

### b)

```python
#Principal Component analysis
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
comp = pca.fit_transform(data_ss)

df_pca = pd.DataFrame(data=comp, columns=['PC1', 'PC2'])
df_pca['target'] = df['target']

df_pca
```

### c)

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['target'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar(label='Target')
plt.show()
```

# Q2)

```python
dfw = pd.read_csv('/content/wine_data.csv')
```

```python
dfw
```

```python
dfw['class_label'].unique()
```

```python
dfw.describe()
```

```python
dfw_without_label = dfw.drop(columns='class_label')
```

```python
col = dfw_without_label.columns
```

```python
from sklearn.preprocessing import StandardScaler


ss = StandardScaler()
dfw1 = ss.fit_transform(dfw_without_label)
dfws = pd.DataFrame(dfw1,columns=col)
dfws
```

### a)

```python
# Boxplot for outlier analysis
plt.figure(figsize=(26,10))
dfws.boxplot()
plt.title('Boxplot of Wine Data Features')
plt.show()
```

```python
# Covariance matrix
cov = dfws.cov()
plt.figure(figsize=(10, 8))
sns.heatmap(cov, annot=True, cmap='coolwarm')
plt.title('Covariance Matrix of Wine Data Features')
plt.show()
```

### b2)

```python
 #Principal Component analysis
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
comp = pca.fit_transform(dfws)
dfw_pca = pd.DataFrame(data=comp, columns=['PC1', 'PC2'])
dfw_pca
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dfw_pca['PC1'], dfw_pca['PC2'], c=dfw['class_label'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Wine Dataset')
plt.colorbar(label='Target')
plt.show()
```

### b1)

```python
pca2 = PCA(n_components=13)
comp = pca2.fit_transform(dfw)
dfw_pca2 = pd.DataFrame(data=comp, columns=['PC1', 'PC2','PC3','PC4', 'PC5','PC6','PC7', 'PC8','PC9','PC10', 'PC11','PC12','PC13'])
dfw_pca2
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(range(0, 13), pca2.explained_variance_ratio_, color='orange')
plt.ylabel('Variance Ratio')
plt.xlabel('Principal Component')
plt.title('Variance Ratio of the Principal Components')
plt.show()
```
